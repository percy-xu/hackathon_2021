```python
plt.figure(figsize = (12, 6))
sns.set_style('whitegrid')
news_real_reply_count_dist = sns.distplot(news_real_reply_count['reply_count'], hist=True)
news_real_reply_count_dist.set(xlim=(0, 20))
news_real_reply_count_dist.set_xlabel("Count", fontsize = 18)
news_real_reply_count_dist.set_ylabel("Probability Density", fontsize = 18)
news_real_reply_count_dist.set_title('Count of Real News Replies', fontsize = 18)
```

```python
# z-score
def normal(x):
    sigma = x.std()
    return x/sigma
```

Density here is the probability density of per unit value, so it can exceed 1.

```python
plt.figure(figsize = (12,6))
news_real_reply_count['normalized_count'] = normal(news_real_reply_count['reply_count'])
news_real_reply_count_dist2 = sns.distplot(news_real_reply_count['normalized_count'], hist=True)
news_real_reply_count_dist2.set(xlim=(0, 1))
news_real_reply_count_dist2.set_xlabel("Normalized Count", fontsize = 18)
news_real_reply_count_dist2.set_ylabel("Per unit value Probability Density", fontsize = 18)
news_real_reply_count_dist2.set_title('Normalized Count of Real News Replies', fontsize = 18)
plt.show()
```
